```python
# Amortization
p= 240000.00
r=0.055
t=30
numerator=(r/12)*(1+(r/12))**(12*t)
denominator=(1+(r/12))**(12*t)-1
payment=round(p*numerator/denominator,2)
print("monthly payment= ",payment)

balance=p
print("month \t balance \t interest")
for a in range(12*t):
    interest=round(balance*r/12,2)
    if a%24==0 or a==359:
        print(a,"\t",balance,"\t",interest)
    balance=round(balance+interest-payment,2)
```

```python
import math
a=.0005

exp=math.floor(math.log(a,10))
n=round(a/(10**exp),2)

if n>=10:
    exp=exp+1
    n=a/(10**exp)

print(a,"=",n,"*10**",exp)
```

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
xmin=-10
xmax=10
ymin=-100
ymax=100
def f(a,b):

    points=10*(xmax-xmin)
    x=np.linspace(xmin,xmax,points)
    fig,ax=plt.subplots()
    plt.axis([xmin,xmax,ymin,ymax])
    plt.plot([xmin,xmax],[0,0])
    plt.plot([0,0],[ymin,ymax])
    plt.plot(x,a*b**x)
    plt.show()
interactive_plot=interactive(f,a=(1,9),b=(1,9))
interactive_plot
```

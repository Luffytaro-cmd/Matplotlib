```python
people = {
      "first" : ['Badal','Dibya','Basu'],
      "last" : ['Dahal','Gurung','Gurung'],
      "email" : ['dahalbadal@7gmail.com', 'dibyagrg494@gmail.com','basugrg@gmail.com']
}
people["email"]
import pandas as pd
df = pd.DataFrame(people)
df.columns = ['first_name','last_name','email']
df.columns
df

df.columns = [x.lower() for x in df.columns]
df

df.columns = df.columns.str.replace('_',' ')
df

df.rename(columns={'first name':'first','last name':'last'},inplace=True)
df

df.loc[2] = ['Basu','Dahal','basudahal@gmail.com']
df

def update_email(email):
    return email.upper()
df['email'].apply(update_email)
df['email'] = df['email'].apply(update_email)
df

df['email'] = df['email'].apply(lambda x:x.lower())
df

df.apply(pd.Series.min)

df.applymap(len)

df.applymap(str.lower)

df['first'].map({'Badal':'dibya','Dibya':'badal'})

df['first']. replace ({'Badal':'dibya','Dibya':'badal'})
```

```python
people = {
      "first" : ['Badal','Dibya','Basu'],
      "last" : ['Dahal','Gurung','Gurung'],
      "email" : ['dahalbadal@7gmail.com', 'dibyagrg494@gmail.com','basugrg@gmail.com']
}
people["email"]
import pandas as pd
df = pd.DataFrame(people)

#adding and removing columns and rows

df['first']+' '+df['last']
df['full_name'] = df['first']+' '+df['last']
df

df.drop(columns=['first','last'],inplace=True)
df

df['full_name'].str.split(' ',expand=True)
df[['first','last']] = df['full_name'].str.split(' ',expand=True)
df

df._append({'first':'Tony'},ignore_index=True)

people = {
      "first" : ['Tony','Steve'],
      "last" : ['Stark','Rogers'],
      "email" : ['ironman.avenge','cap@avenge']
}
df2 = pd.DataFrame(people)
df2
df._append(df2,ignore_index=True)

df=df._append(df2,ignore_index=True)
df

#sorting
df.sort_values(by='last', ascending=False)
df.sort_values(by=['last','first'], ascending=[False,True])
```

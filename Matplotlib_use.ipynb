```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
plt.style.use('fivethirtyeight')

data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/refs/heads/master/Python/Matplotlib/02-BarCharts/data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
langauge_counter = Counter()
for response in lang_responses:
    langauge_counter.update(response.split(';'))
languages = []
popularity = []
for item in language_counter.most_common(15):

    language.append(item[0])
    popularity.append(item[1])
plt.bar(languages, popularity)
```

```python
#pie chart
import matplotlib.pyplot as plot
plt.style.use('fivethirtyeight')
slices = ['59219','55466','47544','36443','37915']
labels = ['Javascript','SQL','Python','Java','Go']
explode = [0,0,0.1,0,0]
plt.pie(slices,labels=labels,explode=explode,shadow=True,autopct='%1.1f%%',wedgeprops={'edgecolor':'black'})
plt.title("My pie chart")
plt.legend()
plt.tight_layout()
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
plt.plot(ages,dev_salaries,color='#444444',linestyle='--',label='All Devs')
plt.plot(ages,py_salaries,label='python')
overall_median = 52787
plt.fill_between(ages,py_salaries,overall_median,where=(py_salaries>overall_median),interpolate=True,alpha=0.25,label='above avg')
plt.fill_between(ages,py_salaries,overall_median,where=(py_salaries<=overall_median),interpolate=True,alpha=0.25,label ='below avg')
plt.legend()
plt.title('median salary(USD) by age')
plt.xlabel('Ages')
plt.ylabel('median salary(USD)')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
plt.style.use('fivethirtyeight')
dates = [datetime(2019,5,24),
         datetime(2019,5,25),
         datetime(2019,5,26),
         datetime(2019,5,27),
         datetime(2019,5,28),
         datetime(2019,5,29),
         datetime(2019,5,30),
        ]
y = [0,1,3,4,6,5,7]
plt.plot_date(dates,y,linestyle='solid')
plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%b,%d,%Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
from itertools import count
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []
index = count()

fig, ax = plt.subplots()  # Create a figure and an axes object

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    ax.cla()  # Clear the current axes
    ax.plot(x_vals, y_vals)

anim = FuncAnimation(fig, animate, interval=1000, cache_frame_data=False) # Explicitly disable caching

plt.tight_layout()
plt.show()
```
